```python
import glob
import numpy as np
from PIL import Image
import os
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
image_folders = glob.glob('data/asl_alphabet_train/asl_alphabet_train/*')
image_list = glob.glob('data/asl_alphabet_train/asl_alphabet_train/**/*.jpg', recursive=True)
```

```python
labels = sorted([os.path.basename(folder) for folder in image_folders])
```

# Plot class distribution

```python
list_count = {}
for folder in image_folders:
    images_list = len(glob.glob(folder + '/*.jpg'))
    label = os.path.basename(folder)
    list_count[label] = images_list
```

```python
keys = list(list_count.keys())
vals = list(list_count.values())

plt.figure(figsize=(15, 10))
plt.title('Class Distribution')
plt.bar(keys, vals)
```

```python
size_list = [Image.open(image).size for image in image_list]
np.unique(size_list) # All image have a size of 200 x 200
```

```python
count = 0
row, column = 7, 5
plt.figure(figsize=(15, 10))
for folder in sorted(image_folders):
    images = glob.glob(folder + '/*.jpg')
    rand_img = np.random.choice(images)
    label = os.path.basename(folder)
    
    img = Image.open(rand_img)
    count += 1
    plt.subplot(row, column, count)
    plt.title(label)
    plt.imshow(np.array(img))
    
plt.tight_layout()
```
